```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Define the line parameters
w1, w2 = 2, -1  # weights
b = -3          # bias

# Create points for the line
x1 = np.linspace(-5, 5, 100)
# From w1*x1 + w2*x2 + b = 0, solve for x2
x2 = -(w1*x1 + b)/w2

# Create the plot
plt.figure(figsize=(10, 8))
plt.plot(x1, x2, 'b-', label='Decision Boundary')

# Add arrows to show the positive and negative regions
plt.arrow(0, 0, w1, w2, head_width=0.2, head_length=0.2, fc='red', ec='red', label='Weight vector')

# Add some styling
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('Decision Boundary: $w_1x_1 + w_2x_2 + b = 0$')

# Add text to show positive and negative regions
plt.text(3, 3, 'z > 0', fontsize=12)
plt.text(-3, -3, 'z < 0', fontsize=12)

plt.legend()
plt.axis('equal')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

plt.arrow(0, 0, w1, w2, head_width=0.2, head_length=0.2, fc='red', ec='red', label='Weight vector (2, -1)')

# Add weight vectors pointing in different directions
# plt.arrow(0, 0, 2, 1, head_width=0.2, head_length=0.2, fc='green', ec='green', label='Weight vector (2, 1)')
# plt.arrow(0, 0, -1, -2, head_width=0.2, head_length=0.2, fc='purple', ec='purple', label='Weight vector (-1, -2)')
# plt.arrow(0, 0, 0, 2, head_width=0.2, head_length=0.2, fc='orange', ec='orange', label='Weight vector (0, 2)')


# # Original red weight vector and its boundary
# w1, w2 = 2, -1  # weights
# b = -3          # bias
# x1 = np.linspace(-5, 5, 100)
# x2 = -(w1*x1 + b)/w2
# plt.plot(x1, x2, 'r-', label='Red Boundary')
# plt.arrow(0, 0, w1, w2, head_width=0.2, head_length=0.2, fc='red', ec='red', label='Weight vector (2, -1)')

# # Green weight vector and its boundary
# w1_g, w2_g = 2, 1
# b_g = -3
# x2_g = -(w1_g*x1 + b_g)/w2_g
# plt.plot(x1, x2_g, 'g-', label='Green Boundary')
# plt.arrow(0, 0, w1_g, w2_g, head_width=0.2, head_length=0.2, fc='green', ec='green', label='Weight vector (2, 1)')

# # Purple weight vector and its boundary
# w1_p, w2_p = -1, -2
# b_p = -3
# x2_p = -(w1_p*x1 + b_p)/w2_p
# plt.plot(x1, x2_p, 'purple', label='Purple Boundary')
# plt.arrow(0, 0, w1_p, w2_p, head_width=0.2, head_length=0.2, fc='purple', ec='purple', label='Weight vector (-1, -2)')

# # Orange weight vector and its boundary
# w1_o, w2_o = 0, 2
# b_o = -3
# x2_o = -(w1_o*x1 + b_o)/w2_o  # This will be a vertical line since w1=0
# plt.axvline(x=b_o/w2_o, color='orange', label='Orange Boundary')
# plt.arrow(0, 0, w1_o, w2_o, head_width=0.2, head_length=0.2, fc='orange', ec='orange', label='Weight vector (0, 2)')



plt.show()
```
